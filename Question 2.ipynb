### Question 2: Do countries with less religious populations have higher education levels?
#### Which ten countries have the smallest religious population?
#### Which ten countries have the highest education levels?
#### Are these countries intersect with each other, if so how many?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stb

edu_countries = pd.read_excel("EdStatsCountry.xls")
edu_countries_series = pd.read_excel("EdStatsCountry-Series.xls")
edu_data2 = pd.read_csv("EdStatsData.csv")
edu_footnote = pd.read_excel("EdStatsFootNote.xls")
edu_series = pd.read_excel("EdStatsSeries.xls")

religion_global = pd.read_excel("global.xls")
religion_national = pd.read_excel("national.xls")
religion_regional = pd.read_excel("regional.xls")
```

#### Are these countries intersect with each other, if so how many?

#### Intersection

#### Year 2010

```python
low_rel_pop_10 = religion_national[religion_national.year==2010][['state','code', 'religion_sumpercent']].sort_values(by='religion_sumpercent', ascending=True)[:20]
lar_ed_lev_10 = edu_data2[edu_data2["Indicator Code"] == "BAR.NOED.15UP.ZS"][['Country Name', 'Country Code', 'Indicator Code', '2010']].sort_values(by='2010', ascending=True)[:20]


set(low_rel_pop_10["code"].values) & set(lar_ed_lev_10["Country Code"].values)
```

#### Year 2005

```python
low_rel_pop_05 = religion_national[religion_national.year==2005][['state','code', 'religion_sumpercent']].sort_values(by='religion_sumpercent', ascending=True)[:20]
lar_ed_lev_05 = edu_data2[edu_data2["Indicator Code"] == "BAR.NOED.15UP.ZS"][['Country Name', 'Country Code', 'Indicator Code', '2005']].sort_values(by='2005', ascending=True)[:20]


set(low_rel_pop_05["code"].values) & set(lar_ed_lev_05["Country Code"].values)
```

#### Hypothesis Testing using total religion sum percent and education data for 2010

```python
set_1 = religion_national[religion_national.year==2010]['religion_sumpercent']
set_2 = edu_data2[edu_data2["Indicator Code"] == "BAR.NOED.15UP.ZS"]['2010']

stb.f_oneway(set_1, set_2.dropna().values)
```
